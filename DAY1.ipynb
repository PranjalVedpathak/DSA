```python
## Divide and Conquer ##
arr1 = [2,4,6,23,45,32,56,75,98,67,27]

def MinMax(arr):
    
    # Base case: only one element
    if len(arr) == 1:
        return arr[0], arr[0]
    
    # Base case: two elements
    if len(arr) == 2:
        if arr[0] < arr[1]:
            return arr[0], arr[1]
        else:
            return arr[1], arr[0]
    
    # Divide
    mid = len(arr) // 2
    left_min, left_max = MinMax(arr[:mid])
    right_min, right_max = MinMax(arr[mid:])
    
    # Conquer (combine results)
    if left_min < right_min:
        final_min = left_min
    else:
        final_min = right_min

    if left_max > right_max:
        final_max = left_max
    else:
        final_max = right_max

    return final_min, final_max
    #return min(left_min, right_min), max(left_max, right_max)


minimum, maximum = MinMax(arr1)
print("Minimum:", minimum)
print("Maximum:", maximum)
```
